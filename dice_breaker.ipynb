```python
# The purpose of this program is to use the random module in python 
# to roll a dice for our icebreaker today. Korren is going to attempt
# to do this live, be prepared for typos.

# import our random library
import random

# declare our variables to determine the smallest and the largest values we want
min = 1
max = 6

# declare a variable to keep track of if we want to roll again, set it to yes to start
roll_again = "yes"

# start our loop
while roll_again == "yes":
    # print message to indicate that the die is rolling
    print("The die is rolling...")
    # declare a varible to assign our random number
    die_value = (random.randint(min,max))
    
    # print the value of the die_value varible
    print("The die value is: ", die_value)
    
    # create our if statement to tell the program what question to ask
    if die_value == 1:
        print("If you were invisible for a day what would you do?")
    elif die_value == 2:
        print('If you could choose a super power, what would it be?')
    elif die_value == 3:
        print("What is the grossest food you have ever tried?")
    elif die_value == 4:
        print("Who is your favourite book or movie character?")
    elif die_value == 5:
        print("What is your favourite game to play?")
    else:
        print("What is one thing you are really good at?")
        
    # prompt the user to roll again
    roll_again = input("Roll again? yes/no")
    
    # Ending the program and error handling code
    if roll_again == "no":
        print("you have selected no, program terminating")
    elif roll_again != "yes" or roll_again != "no":
        print("invalid entry, program ending")

print("This message will only display at the end")
```
